```python
#Record a video with cv2 VideoWriter

import numpy as np
import cv2

cap = cv2.VideoCapture(0)

width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'DIVX')
out = cv2.VideoWriter('output.avi',fourcc, 20.0, (width,height))

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        frame = cv2.flip(frame,0)

        # write the flipped frame
        out.write(frame)

        cv2.imshow('frame',frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

# Release everything if job is finished
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
#Open ctest video in gray

import numpy as np
import cv2
import time

cap = cv2.VideoCapture('output.avi')

if cap.isOpened()==False:
    print("check your file path")

while(cap.isOpened()):
    ret, frame = cap.read()
    frame = cv2.flip(frame,0)
    
    #we wrote the video on 20 fps
    #so we need a delay of 1/20 if we want to see it(else video will play in speed)
    time.sleep(1/25)
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):#0x77 == 27 for esc
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#Show a video with cv2 and draw recangle on it

import numpy as np
import cv2

cap = cv2.VideoCapture(0)

width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#Top left corner
#// to return int value

x=width//2
y=height//2

#Bottom right corner OR with and ht of rectangle

w=width//4
h=height//4



while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        
        cv2.rectangle(frame,(x,y),(w,h),(255,255,0),5)
        
        cv2.imshow('frame',frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

# Release everything if job is finished
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
#Interactive rectangle on video

import numpy as np
import cv2

def draw_rectangle(event,x,y,flags,param):
    
    
    global pt1,pt2,top_left_clicked,bottom_right_clicked
    
    if event==cv2.EVENT_LBUTTONDOWN:
        
        #Reset the rectangle
        
        if top_left_clicked==True and bottom_right_clicked==True:
            
            #give 0 value to pt1 and 2
            pt1=(0,0)
            pt2=(0,0)
            
            #Give false value to tracker
            
            top_left_clicked=False
            bottom_right_clicked=False
    
    #Check top left clickedd if it is false
    
    if top_left_clicked==False:
        pt1=(x,y)
        top_left_clicked=True
        
    elif bottom_right_clicked==False:
        pt2=(x+80,y+80)
        bottom_right_clicked=True

#Global vars        
pt1=(0,0)
pt2=(0,0)    

#Tracker
top_left_clicked=False
bottom_right_clicked=False       
        

cap = cv2.VideoCapture(0)

#Create named window for connectors
cv2.namedWindow("Test")



# mouse Callback
cv2.setMouseCallback("Test",draw_rectangle)


while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        
        #Based o global variable draw frame
        
        if top_left_clicked==True:
            cv2.circle(frame,pt1,25,(0,255,0),2)
            
        if top_left_clicked==True and bottom_right_clicked==True:
            cv2.rectangle(frame,pt1,pt2,(255,0,0),-1)
        
     
        
        cv2.imshow('Test',frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

# Release everything if job is finished
cap.release()
#out.release()
cv2.destroyAllWindows()
```
